Data Analysis with Pandas

```python
import pandas as pd
import numpy as np
```

```python
#A Pandas program to compare the elements of two Pandas series. 
#Sample Series [4,65,436,3,9], [7,0,3,897,9]
s1 = pd.Series([4, 65, 436, 3, 9])
s2 = pd.Series([7, 0, 3, 897, 9])
# Compare the two series
print("Equal: \n", s1 == s2)
print("Not Equal: \n", s1 != s2)
print("Greater than: \n", s1 > s2)
print("Less than: \n", s1 < s2)
print("Greater than or equal to: \n", s1 >= s2)
```

```python
#a Pandas program to add, subtract, multiply and divide two Pandas series. 
#Sample Series: [2,4,6,8,14], [1,3,5,7,9]
s1 = pd.Series([2, 4, 6, 8, 14])
s2 = pd.Series([1, 3, 5, 7, 9])
print("Addition: \n", s1 + s2)
print("Subtraction: \n", s1 - s2)
print("Multiplication: \n", s1 * s2)
print("Division: \n", s1 / s2)
```

```python
#a Pandas program to convert a dictionary to a Pandas series. 
#Sample dictionary: dictionary1 = {‘Josh’: 24, ‘Sam’: 36, ‘Peace’: 19, ‘Charles’: 65, ‘Tom’: 44}
dictionary1 = {'Josh': 24, 'Sam': 36, 'Peace': 19, 'Charles': 65, 'Tom': 44}
s1 = pd.Series(dictionary1)
print(s1)
```

```python
#a Pandas program to convert a given series to an array.
#Sample series: [‘Love’, 800, ‘Joy’, 789.9, ‘Peace’, True]
s1 = pd.Series(['Love', 800, 'Joy', 789.9, 'Peace', True])
array1 = s1.to_numpy()
print(array1)
```

```python
#a Pandas program to display the most frequent value in the given series and replace everything else as ‘Other’ in the series. (Use the ‘HomeTeamGoals’ column in the dataset provided)
df = pd.read_csv('AfricaCupofNationsMatches.csv')
most_frequent = df['HomeTeamGoals'].mode()[0]
df['HomeTeamGoals'] = df['HomeTeamGoals'].apply(lambda x: x if x == most_frequent else 'Other')
print(df['HomeTeamGoals'])
```

```python
#a Pandas program to read the given csv file.
df = pd.read_csv('AfricaCupofNationsMatches.csv')
```

```python
#a Pandas program to get the first 7 rows of your data frame
print(df.head(7))
```

```python
#a Pandas program to select the ‘HomeTeam’, ‘AwayTeam’, ‘HomeTeamGoals’ and ‘AwayTeamGoals’ columns from your data frame.
selected_columns = df[['HomeTeam', 'AwayTeam', 'HomeTeamGoals', 'AwayTeamGoals']]
print(selected_columns)
```

```python
#a Pandas program to select the rows where Egypt appears.
egypt_rows = df[(df['HomeTeam'] == 'Egypt') | (df['AwayTeam'] == 'Egypt')]
print(egypt_rows)
```

```python
#a Pandas program to count the number of rows and columns of your data frame.
rows, columns = df.shape
print(f"Number of rows: {rows},\nNumber of columns: {columns}")
```

```python
#a Pandas program to select the rows where the ‘Attendance’ is missing.
missing_attendance = df[df['Attendance'].isnull()]
print(missing_attendance)
```

```python
#a Pandas program to select the rows where the ‘HomeTeamGoals’ are between 3 and 6 inclusive.
home_goals_between_3_and_6 = df[(df['HomeTeamGoals'] >= 3) & (df['HomeTeamGoals'] <= 6)]
print(home_goals_between_3_and_6)
```

```python
#a Pandas program to change the “AwayTeamGoals” in the 3rd row to 10.
df.at[2, 'AwayTeamGoals'] = 10
print(df.iloc[2]['AwayTeamGoals'])
```

```python
#a Pandas program to sort the DataFrame first by ‘HomeTeam’ in ascending order, then by ‘HomeTeamScores’ in descending order.
sorted_df = df.sort_values(by=['HomeTeam', 'HomeTeamGoals'], ascending=[True, False])
print(sorted_df[['HomeTeam', 'HomeTeamGoals']])
```

```python
#a Pandas program to get list from DataFrame column headers
column_headers = df.columns.tolist()
print(column_headers)
```

```python
#a Pandas program to append a column of your choice to your DataFrame.
df['NewColumn'] = np.random.randint(1, 100, size=len(df))
print(df.head())
```

```python
#a Pandas program to add 2 rows to your DataFrame.
new_rows = pd.DataFrame({
    'HomeTeam': ['NewTeam1', 'NewTeam2'],
    'AwayTeam': ['AwayTeam1', 'AwayTeam2'],
    'HomeTeamGoals': [1, 2],
    'AwayTeamGoals': [0, 3],
    'Attendance': [1000, 2000]
})
df = pd.concat([df, new_rows], ignore_index=True)
print(df.tail(2))
```

```python
#a Pandas program to change the country ‘Uganda’ to ‘China’ in the ‘AwayTeam’ column of the DataFrame.
df['AwayTeam'] = df['AwayTeam'].replace('Uganda', 'China')
print(df[df['AwayTeam'] == 'China'])
```

```python
#a Pandas program to reset index in your DataFrame.
df.reset_index(drop=True, inplace=True)
print(df.head())
```

```python
#a Pandas program to check whether the ‘Stadium’ column is present in your DataFrame or not.
stadium_present = 'Stadium' in df.columns
print(f"Is 'Stadium' column present? {stadium_present}")
```

```python
#a Pandas program to convert the datatype of the ‘AwayTeamGoals’. (int to float)
df['AwayTeamGoals'] = df['AwayTeamGoals'].astype(float)
print(df['AwayTeamGoals'].dtype)
```

```python
#a Pandas program to remove the last 10 rows from your DataFrame.
df = df[:-10]
print(df.tail(10))
```

```python
#a Pandas program to iterate over rows in your DataFrame.
for index, row in df.iterrows():
    print(f"Row {index}: HomeTeam={row['HomeTeam']}, AwayTeam={row['AwayTeam']}, HomeTeamGoals={row['HomeTeamGoals']}, AwayTeamGoals={row['AwayTeamGoals']}")
```

```python
#a Pandas program to change the order of your DataFrame columns.
new_order = ['AwayTeam', 'HomeTeam', 'AwayTeamGoals', 'HomeTeamGoals', 'Attendance']
df = df[new_order]
```

```python
#a Pandas program to delete DataFrame row(s) whose value is 0 in the ‘HomeTeamGoals’ columns.
df = df[df['HomeTeamGoals'] != 0]
print(df[df['HomeTeamGoals'] == 0])  # Should return an empty DataFrame if successful 
```
